```python
import json
import ast

results = []

with open("results.txt") as results_file:
    for line in results_file:
        results.append(ast.literal_eval(line))

print(len(results))
```

```python
import matplotlib.pyplot as plt

print(results[0])

for statistic in ["score.total", "query.aggregates.results.wrong_answer", "query.aggregates.results.ok"]:
    plt.plot([r[statistic] for r in results])
    plt.ylabel(statistic)
    plt.show()
```
